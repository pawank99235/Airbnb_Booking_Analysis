## 3. ***Data Wrangling***
Now time to explore and analyse our dataset to recognize the value between the given data


### Neighbourhood group with respect to room distribution


````python
# Write your code to make your dataset analysis ready.
df2 = df.groupby('neighbourhood_group')['room_type'].value_counts().unstack()
df2
````

here in the above data,we can see that:

*  **Manhattan** in neighbourhood group have maximum number of listings and highest number of apartments which shows **Manhattan** neighbourhood group is the most preferred and busiest.

* Staten island have the least number of listings as well as less number of apartments in airbnb which shows its not preferred that much in comaprison to other neighbourhood group.



### Maximum nights spend by customers in different room types


````python
max_night = df.groupby(['room_type']).agg({'minimum_nights':"sum"}).rename({'minimum_nights':"maximum_nights"},axis = 1).reset_index()
max_night
````

Here in the above analysis we can see that :

 * Most of the customers preferred to stay in entire home apartment in comaprison to private and shared rooms.

* Shared rooms has the least number of nights spend by the customers as many people prefer to have their own privacy.

* followed by entire home apt,**Private room** has the average results in comparison to shared rooms.



###  3 - Top 5 host based on their turnover

````python
df['price'].describe()
````

Let's replace the price less than 50

````python
df.loc[df['price']<50,'price'] = df.loc[(df['price']> 50) & (df['price']< 80),'price'].median( )
df['price'].describe()
````

now we can see,the minimum price is replaced and now we can find our top 5 hosts based on their turnover is airbnb listings

````python
#finding top 5 host based on their turnover
top_host = df.groupby(['host_name','host_id'])['price'].sum().reset_index()
top_host_1 = top_host.sort_values('price',ascending = False).head(5)
top_host_1
````

From the above data we can infer that:


*   The host Sonder(NYC) has highest turnover among all.
*   The turnover provides productive means of judgement of the business scenario and helps the services to sustain effortlessly and provide better services.



##4 Average Price of different room type with respect to different neighbourhood groups


````python
df4 = df.groupby(["neighbourhood_group","room_type"])["price"].mean().unstack()
df4
````

From the above data we can infer that:


*   The Manhatten Neighbourhood group is most costly among all neighbourhood groups offering the services whereas the Bronx neighbourhood group is the cheapest.
*   The average price for all the rooms available gives a brief insight about the affordability of rooms when compared to other applications offering the same services



##top 10 neighborhood having highest number of apartments
